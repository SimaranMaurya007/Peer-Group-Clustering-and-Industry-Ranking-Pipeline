```python
"""
Comprehensive pipeline for:
- Peer-group clustering (multiple algorithms)
- Industry ranking (composite score + optional supervised ranker)
- Cross-modal fusion (text embeddings / TF-IDF fallback)
- Time-series stub for panel forecasting
- Stubs / guidance for advanced models (TabPFN, GNNs, FinBERT)

Save this as pipeline_financial_analysis.py or run in a notebook.
"""

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering, Birch, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from matplotlib import cm
from matplotlib.colors import ListedColormap
import warnings
import seaborn as sns
warnings.filterwarnings("ignore")

# ---------- Config ----------
DATA_PATH = "fda-4_dataset.csv"   # change if needed
OUT_DIR = ""
RANDOM_STATE = 42
MAX_TSNE_POINTS = 20000  # tsne sample size to avoid slowdown
# ----------------------------

# 1) Load
if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"{DATA_PATH} not found. Please place your CSV at this path.")
df = pd.read_csv(DATA_PATH)
print("Loaded", DATA_PATH, "shape:", df.shape)

# 2) Auto-detect numeric/ration-like columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
ratio_keywords = ["ratio","margin","roe","roa","debt","asset","current","quick","eps","pe","beta","turnover","growth","profit","ebit","ebitda","oper","roi"]
for c in df.columns:
    if c not in numeric_cols:
        low = c.lower()
        if any(k in low for k in ratio_keywords):
            coerced = pd.to_numeric(df[c], errors='coerce')
            if coerced.notna().sum() > 0:
                df[c] = coerced
                numeric_cols.append(c)
numeric_cols = sorted(set(numeric_cols))
print("Numeric / ratio-like columns detected:", len(numeric_cols))

if len(numeric_cols) == 0:
    raise ValueError("No numeric columns detected; you need financial ratios or numeric fields to proceed.")

X_raw = df[numeric_cols].copy()

# 3) Defensive cleaning: replace inf, winsorize, impute, scale
X_raw = X_raw.replace([np.inf, -np.inf], np.nan)
# winsorize at 1st and 99th percentiles
lower = X_raw.quantile(0.01)
upper = X_raw.quantile(0.99)
X_capped = X_raw.clip(lower=lower, upper=upper, axis=1)
imp = SimpleImputer(strategy="median")
X_imp = pd.DataFrame(imp.fit_transform(X_capped), columns=X_capped.columns, index=X_capped.index)
scaler = RobustScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_imp), columns=X_imp.columns, index=X_imp.index)

# 4) Composite scoring (bucket heuristic)
buckets = {
    "profitability": ["roe","roa","margin","profit","ebit","ebitda","net income","oper","roi"],
    "liquidity": ["current","quick","liquid"],
    "leverage": ["debt","leverage","debt/equity","debt to"],
    "efficiency": ["turnover","asset turnover","inventory","receivable"],
    "growth": ["growth","yoy","cagr"]
}
feature_buckets = {k:[] for k in buckets}
for col in X_scaled.columns:
    col_low = col.lower()
    assigned = False
    for b, keys in buckets.items():
        for k in keys:
            if k in col_low:
                feature_buckets[b].append(col); assigned=True; break
        if assigned: break
    if not assigned:
        if "turn" in col_low or "eff" in col_low:
            feature_buckets["efficiency"].append(col)
        else:
            feature_buckets["profitability"].append(col)

bucket_weights = {"profitability":0.4,"liquidity":0.15,"leverage":-0.2,"efficiency":0.15,"growth":0.1}
bucket_scores = pd.DataFrame(index=X_scaled.index)
for b in feature_buckets:
    cols_b = feature_buckets[b]
    bucket_scores[b] = X_scaled[cols_b].mean(axis=1) if len(cols_b)>0 else 0.0
df["composite_score"] = sum(bucket_scores[b] * bucket_weights.get(b, 0) for b in bucket_scores.columns)

# 5) Clustering (fast + sample-aware)
def cluster_and_eval(X, k_range=range(2,9), sample_for_sil=2000):
    best = None
    for k in k_range:
        methods = {}
        # work on PCA-reduced version for speed (10 comps)
        pca = PCA(n_components=min(10, X.shape[1]), random_state=RANDOM_STATE)
        Xp = pca.fit_transform(X)
        # KMeans
        km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10).fit(Xp)
        methods["KMeans"] = km.labels_
        # GMM on PCA space
        gm = GaussianMixture(n_components=k, random_state=RANDOM_STATE).fit_predict(Xp)
        methods["GMM"] = gm
        # Agglomerative on PCA space
        ag = AgglomerativeClustering(n_clusters=k).fit_predict(Xp)
        methods["Agglomerative"] = ag
        # Birch
        br = Birch(n_clusters=k).fit_predict(Xp)
        methods["Birch"] = br

        # compute silhouette on a sample to speed up
        if Xp.shape[0] > sample_for_sil:
            idx = np.random.RandomState(RANDOM_STATE).choice(Xp.shape[0], sample_for_sil, replace=False)
            Xs = Xp[idx]
        else:
            Xs = Xp
            idx = np.arange(Xp.shape[0])
        mean_sil = np.nan
        sils = []
        for name, lbls in methods.items():
            try:
                if len(set(lbls[idx])) > 1:
                    sil = silhouette_score(Xs, lbls[idx])
                else:
                    sil = np.nan
            except Exception:
                sil = np.nan
            sils.append(sil)
        mean_sil = np.nanmean([s for s in sils if not np.isnan(s)]) if len([s for s in sils if not np.isnan(s)])>0 else np.nan
        if best is None or (not np.isnan(mean_sil) and mean_sil > best[0]):
            best = (mean_sil, k, methods, pca)
    return best

best_sil, best_k, best_methods, best_pca = cluster_and_eval(X_scaled, k_range=range(2,9))
print("Selected cluster k:", best_k, "silhouette:", best_sil)

# pick KMeans labels by default
labels = best_methods["KMeans"]
df[f"cluster_kmeans_{best_k}"] = labels

# Run DBSCAN separately (no k needed)
db = DBSCAN(eps=2.0, min_samples=5).fit(best_pca.transform(X_scaled))
db_labels = db.labels_
df["cluster_dbscan"] = db_labels
print("DBSCAN clusters found:", len(set(db_labels)) - (1 if -1 in db_labels else 0), " (noise =", sum(db_labels==-1), ")")

# 6) Visualization: PCA 2D + t-SNE (sampled)
pca2 = PCA(n_components=2, random_state=RANDOM_STATE)
X_pca2 = pca2.fit_transform(X_scaled)
plt.figure(figsize=(8,5))
plt.scatter(X_pca2[:,0], X_pca2[:,1], c=labels, s=8,
            cmap="Set1", alpha=0.7)
plt.title(f"PCA 2D colored by KMeans (k={best_k})")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.show()

# DBSCAN visualization on PCA2
plt.figure(figsize=(8,5))
plt.scatter(X_pca2[:,0], X_pca2[:,1], c=db_labels, s=8,
            cmap="tab10", alpha=0.7)
plt.title("PCA 2D colored by DBSCAN clusters")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.show()

# t-SNE (sample to MAX_TSNE_POINTS)
n_samples = X_scaled.shape[0]
if n_samples > MAX_TSNE_POINTS:
    sample_idx = np.random.RandomState(RANDOM_STATE).choice(n_samples, MAX_TSNE_POINTS, replace=False)
    X_tsne_in = X_scaled.iloc[sample_idx]
    lbls_tsne = np.array(labels)[sample_idx]
else:
    X_tsne_in = X_scaled
    lbls_tsne = np.array(labels)
tsne_emb = TSNE(n_components=2, random_state=RANDOM_STATE, init="pca", learning_rate="auto").fit_transform(X_tsne_in)
plt.figure(figsize=(8,5))
plt.scatter(tsne_emb[:,0], tsne_emb[:,1], c=lbls_tsne, s=6,
            cmap="Set1", alpha=0.7)
plt.title("t-SNE (sample) colored by cluster (KMeans best)")
plt.show()

# Step 1: Clean dataset (one row per company per year)
# Assuming `company_id` and `year` columns exist
agg_funcs = {col: "mean" for col in numeric_cols}  # take mean for numeric features
agg_funcs["composite_score"] = "mean"              # also aggregate composite

df_clean = df.groupby(["cik", "ticker", "year"], as_index=False).agg(agg_funcs)

print("After cleaning:", df_clean.shape)

# Step 2: Ask user for year
selected_year = int(input("Enter the year for ranking: "))

df_year = df_clean[df_clean["year"] == selected_year].copy()

if df_year.empty:
    raise ValueError(f"No data found for year {selected_year}")

# Recalculate rankings within the selected year
ranked = df_year.sort_values("composite_score", ascending=False).reset_index(drop=True)
print(f"\nTop 10 by composite_score for year {selected_year}:")
display(ranked.head(10))

# === Bucket-wise Rankings (per year) ===
print(f"\n=== Top companies by buckets for {selected_year} ===")
bucket_rankings = {}
for b in bucket_scores.columns:
    df_year[f"{b}_score"] = bucket_scores.loc[df_year.index, b]   # slice scores for that year
    ranked_b = df_year.sort_values(f"{b}_score", ascending=False).reset_index(drop=True)
    bucket_rankings[b] = ranked_b
    print(f"\nTop 5 by {b} score ({selected_year}):")
    display(ranked_b.head(5))

# === Individual Metric Rankings (per year) ===
print(f"\n=== Top companies by individual metrics for {selected_year} ===")
col_rankings = {}
key_metrics = ["ROA","ROE","ROI","revenue","ProfitMargin", "NetIncome", "PERatio"]
for metric in key_metrics:
    matches = [c for c in numeric_cols if metric in c.lower()]
    for c in matches:
        ranked_c = df_year.sort_values(c, ascending=False).reset_index(drop=True)
        col_rankings[c] = ranked_c
        print(f"\nTop 5 by {c} ({selected_year}):")
        display(ranked_c.head(5))

print("\nDone.")

```
